```python
import pandas as pd
```

```python
path = './wikihow_japanese/data/output/test.jsonl'
```

```python
test_data = pd.read_json(path, orient='records', lines=True)
```

```python
print(test_data.head)
```

```python
import s_methods
import Rouge
USE_s_1 = 0
USE_5_1 = 0
pytext_s_1 = 0
pytext_5_1 = 0
pytext_sum_1 = 0
USE_s_2 = 0
USE_5_2 = 0
pytext_s_2 = 0
pytext_5_2 = 0
pytext_sum_2 = 0
USE_s_L = 0
USE_5_L = 0
pytext_s_L = 0
pytext_5_L = 0
pytext_sum_L = 0

for i,test in enumerate(test_data['src'][:10]):
    USE_s = s_methods.USE_sent(test)
    USE_5 = s_methods.USE_5w1h(test)
    pytext_s = s_methods.pyrank_sent(test)
    pytext_5 = s_methods.pyrank_5w1h(test)
    pytext_sum = s_methods.pytext_sum(test)
    print(USE_s)
    print('\nUSE_5：{}'.format(USE_5))
    print('\npytext_s：{}'.format(pytext_s))
    print('\npytext_5：{}'.format(pytext_5))
    print('\npytext_sum：{}'.format(pytext_sum))
    print('\n答え：{}'.format(test_data['tgt'][i]))
    USE_s_1 = USE_s_1 +  Rouge.Rouge1(USE_s,test_data['tgt'][i])
    USE_5_1 = USE_5_1 +  Rouge.Rouge1(USE_5,test_data['tgt'][i])
    pytext_s_1 = pytext_s_1 + Rouge.Rouge1(pytext_s,test_data['tgt'][i])
    pytext_5_1 = pytext_5_1 + Rouge.Rouge1(pytext_5,test_data['tgt'][i])
    pytext_sum_1 = pytext_sum_1 + Rouge.Rouge1(pytext_sum,test_data['tgt'][i])
    USE_s_2 = USE_s_2 +  Rouge.Rouge2(USE_s,test_data['tgt'][i])
    USE_5_2 = USE_5_2 +  Rouge.Rouge2(USE_5,test_data['tgt'][i])
    pytext_s_2 = pytext_s_2 + Rouge.Rouge2(pytext_s,test_data['tgt'][i])
    pytext_5_2 = pytext_5_2 + Rouge.Rouge2(pytext_5,test_data['tgt'][i])
    pytext_sum_2 = pytext_sum_2 + Rouge.Rouge2(pytext_sum,test_data['tgt'][i])
    USE_s_L = USE_s_L +  Rouge.RougeL(USE_s,test_data['tgt'][i])
    USE_5_L = USE_5_L +  Rouge.RougeL(USE_5,test_data['tgt'][i])
    pytext_s_L = pytext_s_L + Rouge.RougeL(pytext_s,test_data['tgt'][i])
    pytext_5_L = pytext_5_L + Rouge.RougeL(pytext_5,test_data['tgt'][i])
    pytext_sum_L = pytext_sum_L + Rouge.RougeL(pytext_sum,test_data['tgt'][i])
```

```python
print('USE文ごと_ROUGE-1 : {}'.format(USE_s_1/len(test_data['src'][:5])))
print('USE5w1hごと_ROUGE-1 : {}'.format(USE_5_1/len(test_data['src'][:5])))
print('pytextrank文ごと_ROUGE-1 : {}'.format(pytext_s_1/len(test_data['src'][:5])))
print('pytextrank5w1hごと_ROUGE-1 : {}'.format(pytext_5_1/len(test_data['src'][:5])))
print('pytextrank文refごと_ROUGE-1 : {}'.format(pytext_sum_1/len(test_data['src'][:5])))
print('\n')
print('USE文ごと_ROUGE-2 : {}'.format(USE_s_2/len(test_data['src'][:5])))
print('USE5w1hごと_ROUGE-2 : {}'.format(USE_5_2/len(test_data['src'][:5])))
print('pytextrank文ごと_ROUGE-2 : {}'.format(pytext_s_2/len(test_data['src'][:5])))
print('pytextrank5w1hごと_ROUGE-2 : {}'.format(pytext_5_2/len(test_data['src'][:5])))
print('pytextrank文refごと_ROUGE-2 : {}'.format(pytext_sum_2/len(test_data['src'][:5])))
print('\n')
print('USE文ごと_ROUGE-L : {}'.format(USE_s_L/len(test_data['src'][:5])))
print('USE5w1hごと_ROUGE-L : {}'.format(USE_5_L/len(test_data['src'][:5])))
print('pytextrank文ごと_ROUGE-L : {}'.format(pytext_s_L/len(test_data['src'][:5])))
print('pytextrank5w1hごと_ROUGE-L : {}'.format(pytext_5_L/len(test_data['src'][:5])))
print('pytextrank文refごと_ROUGE-L : {}'.format(pytext_sum_L/len(test_data['src'][:5])))
```
